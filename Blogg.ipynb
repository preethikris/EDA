#### Delivery Performance Analysis

```python
#What is the average delivery time?
order['delivery_time'] = (
    order['order_delivered_customer_date'] - order['order_purchase_timestamp']
).dt.days
order['delivery_time'].mean()
```

```python
#Which regions have the longest delivery time?
ord_cust = order.merge(customer, on='customer_id')
ord_cust['delivery_days'] = (
    ord_cust['order_delivered_customer_date'] -
    ord_cust['order_purchase_timestamp']
).dt.days
ord_cust.groupby('customer_state')['delivery_days'].mean().sort_values(ascending=False).head(1)
```

```python
#How does delivery time affect review scores?
oo=ord_cust.merge(order_review,how='inner',on='order_id')
oo.groupby('review_score')['delivery_days'].mean().sort_index()
```

```python
oo[['review_score','delivery_days']].corr()
```
