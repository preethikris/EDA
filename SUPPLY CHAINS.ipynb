```python
#Price segmentation
#Premium vs budget products

q1=df['Price'].quantile(0.25)
q3=df['Price'].quantile(0.75)
def segment(price):
    if price <=q1:
        return "Budget"
    elif price >= q3:
        return "Premium"
    else:
        return "Standard"
df["Price_segment"] = df["Price"].apply(segment) 
print(df["Price_segment"].value_counts())
```
